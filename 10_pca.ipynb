# Redukcja wymiarowości

Metody redukcji wymiarowości transformują dane wielowymiarowe (o wielu zmiennych) do przestrzeni niskowymiarowej, przy jak najwierniejszym zachowaniu istotnych informacji zawartych w danych.
* wydobycie istotnych cech, redukcja szumu - przetransformowane dane można wykorzystać do budowy skuteczniejszych modeli klasyfikacji lub regresji 
* kompresja danych, redukcja kosztu obliczeniowego i pamięciowego algorytmów
* wizualizacja danych  (mapowanie do 2D lub 3D)
* ogranicza tzw. **przekleństwo wymiarowości** - w wielowymiarowej przestrzeni dane są rzadkie, co zmniejsza skuteczność działania wielu metod statystycznych

Wybrane metody 
* Analiza składowych głównych (PCA)
* Analiza dyskryminacyjna (LDA)
* Skalowanie wielowymiarowe (MDS)

Przykład zastosowania:
* Wizualizacja 2D, 3D
* Klasyfikacja na danych przetworzonych przez PCA
* Klasyfikacja twarzy (zdjęć) za pomocą cech PCA 

## Analiza głównych składowych (PCA)

PCA obraca układ współrzędnych tak, aby maksymalizować w pierwszej kolejności wariancję pierwszej współrzędnej, następnie wariancję drugiej współrzędnej itd.. W ten sposób konstruowana jest nowa przestrzeń obserwacji, w której najwięcej zmienności posiadają początkowe czynniki. 

<div>

</div>

Metoda polega na znalezieniu wartości własnych i wektorów własnych macierzy kowariancji danych. Wektor własny o największej wartości własnej określa kierunek (składową) PCA o największej wariancji. Gdy wybierzemy $k$ składowych o największych wartościach własnych to mogą one posłużyć do rzutowania danych $n$-wymiarowych do przestrzeni $k$-wymiarowej (gdzie $k<n$)

## Algorytm PCA:
1. wycentrowanie danych (wyśrodkowanie wartości dla każdej zmiennej)  
   $$ x_{ij} \leftarrow x_{ij} - \frac{1}{n} \sum_{j=1}^n x_{ij}$$ 
2. wyznaczenie macierzy kowariancji
  $$ \mathbf{R}_{x} = \frac{1}{n} \mathbf{X}\mathbf{X}^T  \quad \text{gdzie} \quad \mathbf{X} = \{ x_{ij}\}$$
3. rozkład macierzy kowariancji na wektory i wartości własne, np.  korzystając z rozkładu SVD 
$$ \mathbf{R}_{x} \mathbf{w}_i = \lambda_i \mathbf{w}_i \qquad i=1,\ldots, N $$
$$\mathbf{S} = \mathbf{W}^{-1}\mathbf{R}_{x}\mathbf{W} $$
gdzie $\mathbf{W}$ zawiera wektory własne a $\mathbf{S}$ wartości własne na diagonali   
4. wybór $k$ składowych o największych wartościach własnych
5. utworzenie macierzy rzutowania $\mathbf{W}_k$ z wybranych $k$ wektorow własnych
6. przekształcenie danych do $k$ wymiarowej przestrzeni $ \mathbf{Y}= \mathbf{W}_k \mathbf{X}  $

## Rekonstrukcja

Transformacja PCA wektora $\mathbf{x}$ z przestrzeni $n$-wymiarowej do $k$-wymiarowej

$$ \mathbf{y} = \mathbf{W} \mathbf{x} $$ 

Rekonstrukcja wektora $\mathbf{y}$ do pierwotnej przestrzeni $n$-wymiarowej
$$ \hat{\mathbf{x}} = \mathbf{W}^T \mathbf{y} $$

Jeżeli $k < n$ to jest to rekonstrukcja stratna. Im więcej składowych $k$ uwzględnimy tym wierniejsze odwzorowanie. 

## Zbiór danych Wine

Znajdźmy składowe główne PCA dla danych Wine, które określone sa przez 13 zmiennych

```python
import pandas as pd

df_wine = pd.read_csv('dane/wine.data', header=None)

df_wine.columns = ['Etykieta klas', 'Alkohol', 'Kwas jabłkowy', 'Popiół', 
'Zasadowość popiołu', 'Magnez', 'Całk. zaw. fenoli', 
'Flawonoidy', 'Fenole nieflawonoidowe', 'Proantocyjaniny', 
'Intensywność koloru', 'Odcień', 'Transmitancja 280/315 nm', 'Prolina']

df_wine.head()
```

```python
X = df_wine.iloc[:, 1:].values
y = df_wine.iloc[:, 0].values
print('Rozmiar danych: ', X.shape)
```

## Centrowanie i standaryzacja danych

Metoda PCA wymaga wyśrodkowania danych (umieszczenie środka układu współrzędnych w centrum rozkładu danych).
Metoda PCA jest wrażliwa na skale, w których wyrażone są wartości zmiennych, dlatego
zastosujmy transformację standaryzacji, która wycentruje dane i przeskaluje zakresy wartości do zbliżonych wartości (odchylenie standardowe równe 1). 

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
X_std = s.fit_transform(X)

np.set_printoptions(suppress=True, precision=3)
print('Średnia: ', np.mean(X_std, axis=0))
print('Odch. std: ', np.std(X_std, axis=0))
```

## Rozkład macierzy kowariancji na wektory i wartości własne

Użyjemy funkcji [numpy.linalg.eig](https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html) do wyznaczenia wartości i wektorów własnych macierzy kowariancji 

```python
import numpy as np

# macierz kowariancji
cov_mat = np.cov(X_std.T)

eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)

print('Wartości własne\n%s' % eigen_vals)
```

Wartość własna odpowiada wariancji zmiennej wzdłuż odpowiadającemu wektorowi własnemu

```python
## Wariancja wzdłuż pierwszej skladowej składowej PCA
x0 = np.dot( X_std, eigen_vecs.T[0])
print(np.var(x0))
```

## Wybór składowych głównych

Posortujemy składowe względem wartości własnych (wariancji) od największej do najmniejszej i wybierzmy $k=2$ pierwszych słodowych

```python
sort_idx = np.argsort(eigen_vals)[::-1]
eigen_vals_sort = eigen_vals[sort_idx]
# ustawiamy wektory wlasne w tym samym porzadku
eigen_vecs_sort = eigen_vecs[:, sort_idx]

import matplotlib.pyplot as plt 
plt.bar(range(1,14), eigen_vals_sort)
plt.title('Wartości własne')
plt.show()
```

```python
# dwie piersze składowe
k=2
# macierz transfroamacji do przestrzeni k-wymiarowej
W = eigen_vecs_sort[:, :k]
print(W)
plt.bar(range(1,14), np.abs(W[:,0]))
plt.title('Wartości absolutne pierwszej składowej PC1');
```

Zmienne dla których wartości składowych są bliskie 0 (tutaj zmienna nr 3) mogą być interpretowane jako nieistotne (najmniej przyczyniają się do wariancji w tej składowej).

## Transformacja do nowej przestrzeni

$$ \mathbf{Y} = \mathbf{W} \mathbf{X}$$

```python
X_pca = X_std.dot(W)
print(X_pca.shape)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1])

plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

PCA jest metodą nienadzorowaną (nie uwzględnia informacji o klasach) ale może posłużyć do wizualizacji danych, które posiadają przypisane etykiety klas. W wielu przypadkach transformacja PCA zachowuje istotne niezbędne informacje dotyczące klas, więc znajduje zastosowanie również jako wstępna transformacja modeli nadzorowanych, np. klasyfikacji.  

Zobaczmy jak wygląda rozkład klas po transformacji PCA

```python
for l in np.unique(y):
    plt.scatter(X_pca[y == l, 0], X_pca[y == l, 1], label=l)

plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower left')
plt.show()
```

## Procent wyjaśnianej wariancji


Procent wyjaśnianej wariancji przez składową $i$-tą wynosi 
$$  m_i = \frac{\lambda_i}{\sum \lambda_j} $$
gdzie $\lambda_i$ jest wartością własną odpowiadająca $i$-tej składowej PCA

```python
total = sum(eigen_vals_sort)
var_exp = []
for ev in eigen_vals_sort:
    var_exp.append(ev/total)
    
cum_var_exp = np.cumsum(var_exp)
```

```python
import matplotlib.pyplot as plt

plt.bar(range(1, 14), var_exp, alpha=0.5, align='center', label='Wyjaśniana wariancja')
plt.step(range(1, 14), cum_var_exp, where='mid', label='Skumulowana wyjasniana wariancja')
plt.ylabel('Procent wyjaśnianej wariancji')
plt.xlabel('Główne składowe')
plt.legend(loc='best')
plt.show()
```

Współczynnik ten często się interpretuje jako ilość informacji jaka została uwzględniona w składowej głównej. Początkowe składowe (o największej wariancji) są najbardziej informatywne. Jeśli uwzględnimy wszystkie składowe to całkowita informacja będzie zachowana $\sum_{i=1}^N \lambda_i = 1$

## Analiza głównych składowych w  scikit-learn

Skorzystajmy z gotowego algorytmu PCA dostepnego w pakiecie ``scikit-learn``



```python
from sklearn.decomposition import PCA

pca = PCA()
X_std_pca = pca.fit_transform(X_std)

# ilość komponentów
print(pca.n_components_)

# piwrwsza składowa PCA
pca.components_[0,:]
```

Współczynnik wyjaśnianej wariancji


```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.bar(range(1, 14), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, 14), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Współczynnik wyjaśnionej wariancji')
plt.xlabel('Główne składowe')
plt.show()
```

## PCA jako transformacja wstępna przed klasyfikacją

Transformacja PCA jest często wykorzystywana jako etap wstępny redukcji wymiarowości danych przez zastosowaniem innych metod Data Mining, np. metod klasyfikacji.



```python
from sklearn.model_selection import train_test_split

# wydzielmy zb. testowy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=0)

# standaryzacja
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)

# PCA do 2 wymiarow
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_std)

# klasyfikator trenowany na składowych g00łownych
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class='ovr', random_state=1, solver='lbfgs')
lr = lr.fit(X_train_pca, y_train)

# predykcja na nowych danych (ocena poprawnosci)
X_test_std = sc.transform(X_test)
X_test_pca = pca.transform(X_test_std)
y_pred = lr.predict(X_test_pca)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## Liniowa analiza dyskryminacyjna (LDA) 

Zadaniem LDA jest znalezienie takiej podprzestrzeni cech, w której klasy są od siebie jak najlepiej rozdzielone - maksymalizowana jest więc separowalność klas. LDA zakłada, że klasy posiadają rozkład normalny oraz kowariancje w każdej z klas są takie same. Wymusza to wstępną normalizację danych przed przystąpieniem do właściwej procedury. LDA jest metodą nadzorowaną i wymaga etykiet danych, na których będzie się uczyć.

<div>

</div>

Znalezione przez LDA podprzestrzenie cech wykorzystywane są jako klasyfikator liniowy lub nowe wektory bazowe służące do redukcji wymiarowości problemu.



## Kryterium Fishera dla 2 klas
Dla dwóch klas szukamy kierunku $\mathbf{w}$, który maksymalizuje kryterium:
$$
F(\mathbf{w})=\frac{\mathbf{w}^{T} \mathbf{S}_{m} \mathbf{w}}{\mathbf{w}^{T} \mathbf{S}_{w} \mathbf{w}}
$$
gdzie $\mathbf{S}_m$ to macierz rozproszeń międzyklasowych 

$$
\mathbf{S}_{m}=\left(\mu_{1}-\mu_{2}\right)\left(\mu_{1}-\mu_{2}\right)^{T}
$$
a $\mathbf{S}_w$ to macierz rozproszeń wewnątrzklasowych 
$$
\mathbf{S}_{w}=p_{1} \Sigma_{1}+p_{2} \Sigma_{2}
$$

gdzie $\mu_1, \mu_2$ to średnie w klasach, $\Sigma_1, \Sigma_2$ kowariancje w klasach

Szukamy takiego kierunku $\mathbf{w}$, który maksymalizuje odległość środków klas i minimalizuje rozrzut wektorów w klasach 

## Algorytm LDA :

1. standaryzacja danych  
2. obliczenie $d$-wymiarowego wektora średnich dla każdej klasy 
3. utworzenie międzyklasowej macierzy rozproszenia $\mathbf{S}_M$ oraz wewnątrzklasowej macierzy rozproszenia $\mathbf{S}_W$
4. Wyznaczenie wartości własnych i wektorów własnych macierzy $\mathbf{S}_W^{-1}\mathbf{S}_M$
4. wybór $k$ składowych o największych wartościach własnych
5. utworzenie macierzy $\mathbf{W}_k$ rzutowania z wybranych $k$ wektorów własnych
6. przekształcenie danych do $k$ wymiarowej przestrzeni $ \mathbf{Y}= \mathbf{W} \mathbf{X}  $

## Średnie położenie wektorów dla każdej klasy

```python
mean_vecs = []
labels = np.unique(y_train)

for label in labels:
    mean_vecs.append(np.mean(X_train_std[y_train==label], axis=0))
    print('klasa %s, środek %s\n' %(label, mean_vecs[label-1]))
```

## Międzyklasowa macierz rozproszenia

```python
d = X_train.shape[1] # liczba cech
mean_overall = np.mean(X_train_std, axis=0)

S_M = np.zeros((d, d))
for label, mean_vec in zip(labels, mean_vecs):
    n = X_train[y_train==label, :].shape[0] 
    mean_vec = mean_vec.reshape(d, 1)            #  tworzy wektor kolumnowy
    mean_overall = mean_overall.reshape(d, 1)    # tworzy wektor kolumnowy
    S_M += n * (mean_vec - mean_overall).dot((mean_vec - mean_overall).T)

print('Międzyklasowa macierz rozproszenia: %sx%s' % (S_M.shape[0], S_M.shape[1]))
```

## Wewnątrzklasowa macierz rozproszenia

Obliczmy kowariancję dla wektorów z każdej z klas osobno

```python
S_W = np.zeros((d, d))

for label in labels:
    class_scatter = np.cov(X_train_std[y_train==label].T)
    S_W += class_scatter

print('Wewnątrzklasowa macierz rozproszenia: %sx%s' % (S_W.shape[0], S_W.shape[1]))
```

## Wyznaczenie kierunków dyskryminujących

Rozwiązuje uogólniony problem wartości własnych dla macierzy $S_W^{-1}S_M$:

```python
eigen_vals, eigen_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_M))
```

Uporządkujmy składowe w kolejności malejących wartości własnych

```python
sort_idx = np.argsort(np.abs(eigen_vals))[::-1]
eigen_vals_sort = eigen_vals[sort_idx]
eigen_vecs_sort = eigen_vecs[:, sort_idx]

print('Wartości własne w malejącej kolejności:\n')
for ev in eigen_vals_sort.real:
    print('%10.5f' % ev)     # moga miec czesc urojona, wypiszmy tylko czesc rzeczywista
```

Tylko 2 wartości własne są większe od zera. Metoda LDA dla danych posiadających $K$ klas znajduje co najwyżej $K-1$ kierunków dyskryminacyjnych. Dane Wine posiadają 3 klasy, więc otrzymujemy 2 składowe. 

## Współczynnik rozróżnialności

Podobnie jak dla PCA możemy zdefiniować współczynnik, który określa w tym wypadku miarę "rozróżnialności" klas

```python
%matplotlib inline
total = sum(eigen_vals.real)
discr = []
for ev in eigen_vals_sort.real:
    discr.append( ev / total)

cum_discr = np.cumsum(discr)

plt.bar(range(1, 14), discr, alpha=0.5, align='center', label='Pojedyncza "rozróżnialność"')
plt.step(range(1, 14), cum_discr, where='mid', label='Łączna "rozróżnialność"')
plt.ylabel('Współczynnik "rozróżnialności"')
plt.xlabel('Dyskryminanty liniowe')
plt.legend(loc='best')
plt.show()
```

## Rzutowanie danych do nowej przestrzeni

```python
W = eigen_vecs_sort[:, :2].real
print('Macierz transformacji W:\n', W)
```

```python
# Rzutowanie danych 13 wymiarowych do przestrzeni 2 wymiarowej
X_train_lda = X_train_std.dot(W)

for l in labels:
    plt.scatter(X_train_lda[y_train==l, 0] , X_train_lda[y_train==l, 1] , label=l)

plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.show()
```

## MDS - skalowanie wielowymiarowe

Kolejną metodą wykorzystywaną przy redukcji wymiarów jest skalowanie wielowymiarowe (MDS). Zadaniem algorytmu jest przekształcenie zbioru danych tak, aby wynikowy zbiór posiadał mniej wymiarów, ale zachowane były euklidesowe odległości między punktami. 

<div>

</div>

## Funkcja stresu

Transformacja MDS polega na znalezieniu takiego zbioru wektorów $\mathbf{y}_i$, który minimalizuje funkcję stresu. 


 $$E=\sum_{i<j}^{n}\left(\delta_{i j}-d_{i j}\right)^{2} \qquad \text{gdzie} \qquad  d_{ij} = ||\mathbf{y}_i - \mathbf{y}_j ||$$
 

Bardzo ważne jest, że dla poprawności działania algorytmu, dane muszę być znormalizowane lub ustandaryzowane (dokładnie jak w przypadku chociażby algorytmu k-średnich). 

## MDS w scikit-learn

MDS zaimplementowany jest w module manifold.

Uwaga: metoda MDS nie dostarcza modelu transformacji (brak metody ``transfrom()``). W przypadku nowego zbioru danych należy wykonać ponownie mapowanie MDS z nowymi punktami. 

```python
from sklearn.manifold import MDS

mds = MDS(2, random_state=0, normalized_stress='auto')
projected = mds.fit_transform(X_std)

for c in np.unique(y):
    plt.scatter(projected[y==c, 0], projected[y==c, 1], label=c)

plt.xlabel('MDS 1')
plt.ylabel('MDS 2')
plt.legend(loc='lower right')
plt.show()
```

# Eigenfaces - rozpoznawanie twarzy

Problem: identyfikacja osób na podstawie zdjęć.   

<div>

</div>

Jest to typowy problem klasyfikacji, jednak dane są w postaci zdjęć (macierz pikseli). Etykietami klas są nazwiska osob, zbiór danych zawiera dziesiątki zdjęć każdej z osób.

Metoda Eigenfaces wykorzystuje wektory własne (PCA). Dane wejściowe to zbiór pikseli obrazu o wymiarze WxH. Metoda PCA pozwala zredukować liczbę wymiarów a znalezione składowe główne (zwane twarzami własnymi) posłużą do skonstruowania zbioru treningowego.

Algorytm ten może zostać wykorzystany nie tylko do rozpoznawania twarzy, ale również pisma czy innych zdjęć (np. obiektów na zdjęciu, takich jak budynki, znaki drogowe, rowery, zwierzęta itp.).

Prześledźmy, jak działa rozpoznawanie twarzy metodą Eigenfaces na przykładowym zbiorze zdjęć.

Funkcja pomocnicza do wyświetlania galerii zdjęć

```python
import matplotlib.pyplot as plt

def plot_gallery(images, titles, h, w, n_row=3, n_col=4):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(min(n_row * n_col, len(images))):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
```
